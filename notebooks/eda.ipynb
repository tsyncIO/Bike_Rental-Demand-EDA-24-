1. Skewness: 0.0037
What it means:

Skewness measures the asymmetry of the distribution. A value of 0 indicates a perfectly symmetric distribution (like a normal distribution), while a positive skew means the distribution has a longer tail on the right, and a negative skew means the tail is longer on the left.
In your case, the skewness value is 0.0037, which is close to 0. This indicates that the data is almost symmetrical. The very small value suggests that there is very little skewness and the data is roughly balanced.
Processing implications:

Since the skewness is very close to zero, no transformation is needed for the data from a skewness perspective. The data seems to be normally distributed, so it's ready for most types of analyses and modeling without needing a transformation like a log or square root.
2. Kurtosis: -0.9145
What it means:

Kurtosis measures the peakedness of the distribution. It tells us if the data has outliers or heavy tails.
A kurtosis of 0 would indicate a normal distribution (neither heavy nor light tails).
A positive kurtosis (>0) means the data has heavy tails or more outliers than a normal distribution.
A negative kurtosis (<0) means the data has light tails or fewer outliers than a normal distribution.
In your case, the kurtosis value is -0.9145, which is negative. This suggests that the distribution of temp has lighter tails than a normal distribution, meaning that the data is less likely to have extreme outliers.
Processing implications:

The negative kurtosis suggests that there are fewer extreme outliers in the data. You might not need to worry about extreme values distorting the analysis.
However, it could also indicate that the data is more uniform and doesn't show extreme spikes, which can sometimes be useful in certain models (especially when dealing with rare events).

Create Rolling Mean/rolling window for Demand (count) to Capture Trends
To analyze trends over time, you can create a rolling mean of demand.

Calculating 7-day rolling average for target column - count

```python
train_df_exp['count'].info()
```

1. This sets the datetime column as the index of the DataFrame.
This is crucial for time series data because it allows operations like rolling averages to be done in a time-aware manner.
Once the datetime column becomes the index, operations like rolling averages will take into account the time ordering of the data.


2. df['count'] refers to the column for which you want to compute the rolling average (in this case, the number of bike rentals).
.rolling(window=7) creates a rolling window of size 7, meaning it will compute the average of the last 7 values at each point.
.mean() computes the mean (average) over each 7-day window.
The result is stored in a new column, rolling_count, which holds the 7-day rolling average of the count column.
For each row in the count column: The rolling average will consider the last 7 rows (including the current row) and compute their mean.
For the first 6 rows, the rolling average will be NaN because there are not enough previous rows to form a 7-day window.

3. After the rolling operation, you might want to reset the index back to the original, which is typically the default integer index (0, 1, 2, ...).
.reset_index() will revert the index back to its original state (if you don't want the datetime column as the index anymore).
This step is optional, depending on whether you want datetime to remain as the index or go back to the default integer-based index.


```python
# Setting datetime index for time series operations
train_df_exp = train_df_exp.set_index('datetime')

train_df_exp['rolling_count'] = train_df_exp['count'].rolling(window=7).mean()

# Resetting index back to original
train_df_exp = train_df_exp.reset_index()

train_df_exp['rolling_count'].head(20)
```

Why is this useful?
Smoothing: The rolling average smooths the data, making trends more apparent and reducing the impact of short-term fluctuations.
Time series analysis: By setting the datetime as the index, the rolling function computes averages over time windows, making it ideal for time series data like bike rentals, sales, stock prices, etc.

Group by Date to Analyze Daily Trends

```python
# Group by date and aggregate
daily_totals = (
    train_df_exp
    .groupby(train_df_exp['datetime'].dt.date)  # Grouping by date (year-month-day)
    .agg({'casual': 'sum', 'registered': 'sum', 'count': 'sum'})  # Summing the values
    .reset_index()  # Resetting the index so 'datetime' becomes a column again
)

# Display the aggregated daily totals
print(daily_totals)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting total demand for casual users
plt.figure(figsize=(10, 6))
sns.barplot(x='datetime', y='casual', data=daily_totals, color='blue')
plt.title('Daily Total Demand for Casual Users')
plt.xlabel('Date')
plt.ylabel('Casual Demand')
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting total demand for registered users
plt.figure(figsize=(10, 6))
sns.barplot(x='datetime', y='registered', data=daily_totals, color='orange')
plt.title('Daily Total Demand for Registered Users')
plt.xlabel('Date')
plt.ylabel('Registered Demand')
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting total count of bikes (Casual + Registered)
plt.figure(figsize=(10, 6))
sns.barplot(x='datetime', y='count', data=daily_totals, color='green')
plt.title('Daily Total Bike Demand (All Users)')
plt.xlabel('Date')
plt.ylabel('Total Demand')
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
